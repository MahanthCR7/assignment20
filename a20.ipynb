## **Q1. What is an API? Why are APIs important?**

### **Answer:**

An API (Application Programming Interface) is a set of rules that allows different software applications to communicate with each other.

APIs are important because they enable integration between systems, allow reuse of functionality, and help in building scalable and modular applications.

---

## **Q2. Explain the core principles of RESTful APIs**

### **Answer:**

* **Stateless:** Each request contains all required information
* **Client-Server:** Separation of frontend and backend
* **Uniform Interface:** Consistent API structure
* **Resource-Based:** Everything is treated as a resource (URL)
* **Scalability:** Designed to handle large-scale systems

---

## **Q3. What are the common HTTP methods used in REST APIs and what are their typical uses?**

### **Answer:**

* **GET:** Retrieve data
* **POST:** Create new data
* **PUT:** Update entire resource
* **PATCH:** Update partial resource
* **DELETE:** Remove resource

---

## **Q4. Explain the difference between stateless and stateful REST APIs**

### **Answer:**

* **Stateless:** Server does not store client data; each request is independent
* **Stateful:** Server stores session or client data between requests

REST APIs are generally stateless.
